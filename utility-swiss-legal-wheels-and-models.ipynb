````python
import os
from pathlib import Path

DST = Path("/kaggle/working")
WHEELS = DST / "wheels"
MODELS = DST / "models"
WEAVIATE_BIN = DST / "weaviate_binary"

for d in [WHEELS, MODELS, WEAVIATE_BIN]:
    d.mkdir(parents=True, exist_ok=True)

print("Directories created")
````

````python
import subprocess, sys

# All packages needed across all downstream notebooks
PACKAGES = [
    # Vector DB + search
    "lancedb",
    "tantivy",
    # Embeddings
    "sentence-transformers==5.4.0",
    # Weaviate (kept for compatibility, may remove later)
    "weaviate-client==4.20.5",
]

subprocess.check_call([
    sys.executable, "-m", "pip", "download",
    "--dest", str(WHEELS),
    "--python-version", "3.12",
    "--only-binary=:all:",
] + PACKAGES)

# Count what we got
wheel_count = len(list(WHEELS.glob("*")))
wheel_size = sum(f.stat().st_size for f in WHEELS.rglob("*") if f.is_file())
print(f"\nDownloaded {wheel_count} wheels ({wheel_size/1e9:.2f} GB)")
````

````python
import subprocess, sys

# Install sentence-transformers temporarily to download the model
subprocess.check_call([
    sys.executable, "-m", "pip", "install", "--quiet",
    "--no-index", "--find-links", str(WHEELS),
    "sentence-transformers==5.4.0",
])

from sentence_transformers import SentenceTransformer

MODEL_NAME = "BAAI/bge-m3"
MODEL_PATH = MODELS / "bge-m3"

if not (MODEL_PATH / "config.json").exists():
    print(f"Downloading {MODEL_NAME}...")
    model = SentenceTransformer(MODEL_NAME)
    model.save(str(MODEL_PATH))
    del model
    print(f"Saved to {MODEL_PATH}")
else:
    print(f"Model already exists at {MODEL_PATH}")

model_size = sum(f.stat().st_size for f in MODEL_PATH.rglob("*") if f.is_file())
print(f"Model size: {model_size/1e9:.2f} GB")
````

````python
import urllib.request, tarfile, platform, stat, os

VERSION = "v1.36.0"
machine = platform.machine()
if machine == "x86_64":
    machine = "amd64"

url = f"https://github.com/weaviate/weaviate/releases/download/{VERSION}/weaviate-{VERSION}-{platform.system()}-{machine}.tar.gz"
tar_path = WEAVIATE_BIN / "tmp_weaviate.tgz"
binary_path = WEAVIATE_BIN / "weaviate"

if not binary_path.exists():
    print(f"Downloading Weaviate {VERSION}...")
    urllib.request.urlretrieve(url, tar_path)
    with tarfile.open(tar_path) as tf:
        tf.extract("weaviate", path=WEAVIATE_BIN)
    tar_path.unlink()
    os.chmod(binary_path, binary_path.stat().st_mode | stat.S_IEXEC | stat.S_IXGRP | stat.S_IXOTH)
    print(f"Binary saved: {os.path.getsize(binary_path)/1e6:.1f} MB")
else:
    print(f"Binary already exists: {os.path.getsize(binary_path)/1e6:.1f} MB")
````

````python
import os
from pathlib import Path

DST = Path("/kaggle/working")

print("Output contents:")
for entry in sorted(os.listdir(DST)):
    path = DST / entry
    if path.is_dir():
        size = sum(f.stat().st_size for f in path.rglob("*") if f.is_file())
        count = sum(1 for f in path.rglob("*") if f.is_file())
        print(f"  {entry}/  ({size/1e9:.2f} GB, {count} files)")
    else:
        print(f"  {entry}  ({path.stat().st_size/1e9:.2f} GB)")

total = sum(f.stat().st_size for f in DST.rglob("*") if f.is_file())
print(f"\nTotal: {total/1e9:.2f} GB / 19.5 GB limit")

# Spot checks
checks = [
    ("wheels/lancedb*", "lancedb wheel"),
    ("wheels/tantivy*", "tantivy wheel"),
    ("wheels/sentence*transformers*", "sentence-transformers wheel"),
    ("wheels/weaviate*", "weaviate-client wheel"),
    ("models/bge-m3/config.json", "bge-m3 model"),
    ("weaviate_binary/weaviate", "weaviate binary"),
]
print("\nSpot checks:")
for pattern, label in checks:
    found = list(DST.glob(pattern))
    status = "✓" if found else "✗ MISSING"
    print(f"  {status}  {label}")
````
